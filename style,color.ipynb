```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tip=sns.load_dataset('tips')
```

```python
sns.countplot(tip,x='sex',palette='coolwarm')
plt.show()
```

```python
sns.set_style('dark')
sns.countplot(tip,x='sex')
plt.show()
```

```python
sns.set_style('ticks')
sns.countplot(tip,x='sex',palette='deep')
plt.show()
```

```python
#spine removing
```

```python
sns.countplot(tip,x='sex')
sns.despine()
plt.show()
```

```python
sns.countplot(tip,x='sex')
sns.despine(left=True)
plt.show()
```

```python
#set size
```

```python
plt.figure(figsize=(15,7))
sns.countplot(tip,x='sex')
plt.show()
```
