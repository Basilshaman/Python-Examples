```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.arange(0,11)
y = x ** 2+5
```

```python
plt.plot(x, y,y,x,'r')
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here');
```

```python
# lay out
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
for i in axes:
    i.plot(x,y,'b')
    i.set_xlabel('x')
    i.set_ylabel('y')
    i.set_title('title')
plt.tight_layout()
```

```python
d=[1,2,3,4,5,4,3,2,1]
u=[5,4,3,2,1,3,4,3,2]
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5)#transperncy
ax.plot(x, x+2, color="#8B008B")        
ax.plot(x, x+3, color="#FF8C00") ; 
#use color with code
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

```

```python
# linestype ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')
```

```python
# custom dash
fig, ax = plt.subplots(figsize=(12,6))
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...
```

```python
# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')
```

```python
# marker size and color
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

```python
fig,ax=plt.subplots(figsize=(8,6))
#color and linewidth(lw)
ax.plot(x,x+1,color='b',linewidth=0.25)
ax.plot(x,x+2,color='b',linewidth=0.5)
ax.plot(x,x+3,color='b',linewidth=0.75)
ax.plot(x,x+4,color='b',linewidth=1)
#color,lw and linestyle(ls)
ax.plot(x,x+5,color='r',linewidth=3,ls='-')
ax.plot(x,x+6,color='r',linewidth=3,ls=':')
ax.plot(x,x+7,color='r',linewidth=3,ls='-.')
#with symbel
ax.plot(x,x+8,color='y',linewidth=3,ls=':',marker='s')
ax.plot(x,x+10,color='y',linewidth=3,ls='-.',marker='+')
ax.plot(x,x+9,color='y',linewidth=3,ls='-',marker='o')
ax.plot(x,x+11,color='y',linewidth=3,ls='-',marker='1')
#cerate disign
line, = ax.plot(x, x+12, color="black", lw=1.50)
line.set_dashes([5, 5, 15, 5])
#set symbel size #color use with code
ax.plot(x,x+13,color='#ff99cc',linewidth=1,ls=':',marker='s',markersize=10)
ax.plot(x,x+14,color='#6600cc',linewidth=1,ls=':',marker='o',markersize=10)
#set marker face color
ax.plot(x,x+15,color='#6600cc',linewidth=1,ls=':',marker='o',markersize=10,markerfacecolor='r')
#set mark face edge  and edge color
ax.plot(x,x+15,color='#6600cc',linewidth=1,ls=':',marker='o',markersize=10,markerfacecolor='r',markeredgewidth=5,markeredgecolor='b')
```
