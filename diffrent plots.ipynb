```python
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
tip=sns.load_dataset('tips')
tip
```

```python
tip.head(6)
```

```python
sns.barplot(x='total_bill',y='sex',data=tip)
plt.show()
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tip,estimator=np.std)
plt.show()
```

```python
sns.countplot(x='sex',data=tip)
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=tip,palette='rainbow')
#palette='rainbow' it is color set
plt.show()
```

```python
sns.boxplot(data=tip,palette='rainbow',orient='h')
#orient='h' is change to horizontal shape
plt.show()
```

```python
tip
```

```python
sns.boxplot(x='day',y='total_bill',hue='smoker',data=tip,palette='rainbow')
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=tip,palette='rainbow')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',hue='smoker',data=tip,palette='rainbow')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',hue='smoker',data=tip,split=True,palette='rainbow')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tip,jitter=True)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tip,jitter=False)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tip,jitter=True,hue='sex')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tip,jitter=True,hue='sex',palette='Set1')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tip,jitter=True,hue='sex',palette='Set2')
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill',data=tip, palette="Set1",split=True)

plt.show()
```

```python
#split replace dodge
sns.swarmplot(x='day',y='total_bill',data=tip, palette="Set1",dodge=True)

plt.show()
```

```python
sns.violinplot(x="tip", y="day", data=tip,palette='rainbow')
sns.swarmplot(x="tip", y="day", data=tip,color='black',size=3)
plt.show()
```

```python
sns.factorplot(x="sex", y="day", data=tip,kind='bar')
plt.show()
```

```python
#factor plot replace cat plot
sns.catplot(x="sex", y="day", data=tip,kind='bar')
plt.show()
```
