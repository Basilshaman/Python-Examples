```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
s=sns.load_dataset('tips')
s.head()
```

```python
sns.lmplot(s,x='total_bill',y='tip')
plt.show()
```

```python
sns.lmplot(s,x='total_bill',y='tip',hue='sex',palette='rainbow')
plt.show()
```

```python
#disegin marker
```

```python
sns.lmplot(s,x='total_bill',y='tip',hue='sex',palette='coolwarm',markers=['o','s'],scatter_kws={'s':100})
plt.show()
# kws={'s':100} this 100 is marker size
```

```python
sns.lmplot(s,x='total_bill',y='tip',col='sex',palette='coolwarm')
plt.show()
```

```python
sns.lmplot(s,x='total_bill',y='tip',col='sex',row='time',palette='coolwarm')
plt.show()
```

```python
sns.lmplot(s,x='total_bill',y='tip',col='day',hue='sex',palette='rainbow')
plt.show()
```

```python
sns.lmplot(s,x='total_bill',y='tip',col='day',hue='sex',palette='rainbow',aspect=0.6,height=6)
plt.show()
```
