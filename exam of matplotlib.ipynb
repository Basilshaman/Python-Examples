#Follow along with these steps

Create a figure object called fig using plt.figure()
Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
Plot (x,y) on that axes and set the labels and titles to match the plot below

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,'r')
axes.set_ylabel('y')
axes.set_xlabel('x')
axes.set_title('title')
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,'r')
axes.set_ylim([0,200])
axes.set_xlim([0,100])
axes.set_title('title')
plt.yticks(np.arange(0, max(y)+50, 50))
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) 
axes1.set_title("out")

axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.2]) 
axes2.set_title("in")
axes2.set_yticks(np.arange(0, max(y)+0.2, 0.2))
axes2.set_xticks(np.arange(0,max(y)+0.2, 0.2))
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) 
axes1.set_title("out")
axes1.plot(x,y,'r')
axes1.set_ylim([0,200])
axes1.set_xlim([0,100])

axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.2]) 
axes2.set_title("in")
axes2.plot(x,y,'r')
axes2.set_ylim([0,200])
axes2.set_xlim([0,100])
axes1.set_title('title')
```
