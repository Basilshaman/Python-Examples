```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fli=sns.load_dataset('flights')
fli.head()
```

```python
tip=sns.load_dataset('tips')
tip.head()
```

```python
tip.corr(numeric_only=True)
```

```python
sns.heatmap(tip.corr(numeric_only=True))
plt.show()
```

```python
sns.heatmap(tip.corr(numeric_only=True),cmap='rainbow',annot=True)
plt.show()
```

```python
a=fli.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(a)
plt.show()
```

```python
sns.heatmap(a,cmap='magma',lw=1,color='white')
plt.show()
```

```python
sns.clustermap(a)
plt.show()
```

```python
sns.clustermap(a,cmap='rainbow',standard_scale=1)
plt.show()
```
