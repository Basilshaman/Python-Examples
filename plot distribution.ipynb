```python
import seaborn as sns
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
tip=sns.load_dataset('tips')
tip
```

```python
tip.head()
```

```python
sns.distplot(tip['total_bill'])
plt.show()
```

```python
sns.displot(tip['total_bill'])
plt.show()
```

```python
sns.histplot(tip['total_bill'])
plt.show()
```

```python
sns.histplot(tip['total_bill'],kde=False,bins=50)
plt.show()
```

```python
sns.histplot(tip['total_bill'],kde=True,bins=50)
plt.show()
```

```python
sns.distplot(tip['total_bill'],kde=False,bins=50,hist=True)
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tip,kind='scatter')
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tip,kind='hex')
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tip,kind='reg')
plt.show()
```

```python
sns.pairplot(tip)
plt.show()
```

```python
sns.pairplot(tip,hue='sex',palette='rainbow')
plt.show()
```

```python
sns.rugplot(tip['total_bill'])
plt.show()
```
